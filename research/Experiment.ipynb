```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Encode the first column ('Category')
wines_clean['type'] = label_encoder.fit_transform(wines_clean['type'])
wines_clean = wines_clean.reset_index(drop=True)
wines_clean
```

```python
wines_clean.iloc[510:522,:]
```

```python
from sklearn.preprocessing import StandardScaler



scaler = StandardScaler()
wines_standard = pd.DataFrame(scaler.fit_transform(wines_clean.iloc[:, 1:12]), columns=wines_clean.columns[1:12])
print(wines_standard.columns)

wines_standard.insert(0, 'type', wines_clean['type'])
wines_standard['quality'] = wines_clean['quality']
```

```python
wines_standard.iloc[510:522,:]
```

```python
wines_standard.hist(bins=50, figsize=(20,15), color="orange")
plt.show()
```

```python
rows_with_null = wines_standard[wines_standard.isnull().any(axis=1)]

# Displaying the rows with null values
print("\nRows with null values:")
print(rows_with_null.iloc[15:,:])
```

```python
from sklearn.model_selection import train_test_split

X = wines_standard.drop(["quality"], axis=1)
y = wines_standard["quality"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import ElasticNet

lr = ElasticNet(alpha=0.01, l1_ratio=0.5,random_state=42)
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
import numpy as np
```

```python
def eval_metrics(actual,pred):
 rmse = np.sqrt(mean_squared_error(actual,pred))
 mae = mean_absolute_error(actual,pred)
 r2 = r2_score(actual,pred)

 return rmse,mae,r2
```

```python
predicted_value = lr.predict(X_test)
```

```python
(rmse,mae,r2) = eval_metrics(y_test,predicted_value)
```

```python
print("rmse = ", rmse)
print("mae = ",mae)
print("r2 :",r2)
```
